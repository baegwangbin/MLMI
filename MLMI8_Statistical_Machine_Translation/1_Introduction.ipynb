# 1. Introduction

## 1.1. What is Machine Translation?

* **Machine Translation**

>* Full or partial automation of translation as carried out by humans
>* Translated texts are readily available (e.g. EU Parliamentary Proceedings)

* **Different Tasks in MT**

>* **Translation:** interpretaion of the meaning of a text and the subsequent production of an equivalent text that communicates the same message in another language
>* **Interpretation:** the intellectual activity that consists of facilitating oral or sign-language communication, either simultaneously or consecutively, between two or among three or more speakers who are not speaking, or signing, the same language
>* **Transliteration:** writing a word from one language using the closest corresponding orthography of a different language
>* **Summarization, Glossing, ...**

* **Examples of MT**

>* News Sources with Multilingual and Multinational Audiences
>  * Add additional info. / Remove unnecessary info. / ...
>* Social Media
>  * Short, noisy messages / untranslatable without context / ...
>* Product Localization
>  * Product name should be preserved / ...

## 1.2. Challenges in Machine Translation

* **Domain and Genre**

>* **Sublanguages:** Domain/Genre-specific languages
>* Difference sublanguages differ in:
>  * Specialized vocabularies
>  * Stylistic differences (e.g. active vs. passive)
>  * Sentence length (e.g. weather forecast vs. news stories)

* **Word Sense and Context**

><img src = 'images/image1_01.png' width=500>
>* Ambiguous words should be translated based on their sense, which is often determined by context
>* Idioms may require different word sense, depending on the language

>|Language|Idioms|Word|
|-|-|-|
|English|you're just pulling my **leg**|**leg: leg**|
|Italian|mi stai prendeno per il **naso**|**naso: nose**|
|Spanish|me est tomando el **pelo**|**pelo: hair**|

* **Morphology**

>* Polysynthetic languages (e.g. Siberian): A single word might correspond to a single English sentence
>* Agglutinative languages (e.g. Turkish): Morphemes are segmentable
>* Fusion languages (e.g. Russian): Affixes and prefixes carry syntactic meaning

* **Movement - Variations in Word Order**

>* Example: S-V-O (e.g. English) vs. S-O-V (e.g. Japanese) vs. V-S-O (e.g. Arabic)
>* Compuational cost of reordering grows exponentially with sentence length

* **Pronouns**

>* Drops pronouns (e.g. Japanese)
>* Drops subject pronouns (e.g. Spanish - often indicated from verbal inflection)

* **Tenses**

>* Spanish: two versions of past tense - definite time & unknown time
>* Translation from **English** to **Spanish**: requires selection of tense

* **Transliteration**

>* Closeness is often defined **phonetically**
>* MT system must:
>  * Identify which words should be transliterated
>  * Identify transliterated words and normalize them
>* Transliteration into Chinese: a written form with the given pronunciation is chosen
>  * Multiple transliterations for same word

* **Other Challenges**

>* **Codeswitching** (e.g. "Hasta la vista, baby" $\rightarrow$ "Sayonara, Amigo")
>* **Variability of Human Translations** 
>* **Discourse and Meaning** (e.g. "This sentence has five words")
>* **Grammatical gender**
>* **Translation is Impossible** (e.g. "This is a Gavagi" $\rightarrow$ "What is a Gavagi?")
