# 1. What is Classification?

* **Definition**:

>Class of supervised machine learning tasks in which the aim is to predict a **discrete valued output** \\(y^\star\\) at an **input** location \\(x^\star\\) using a **training set of input-output pairs** \\(\{x_n, y_n\}_{n=1}^N\\) pairs.

* **Goal**:

>1. Return **label** $y^\star$ or **probability distribution** $p(y^\star|x^\star)$
>2. Understand the relationship between the inputs and the outputs (e.g. parameters that affect voting results)

* **Comparison to Regression**:

>* Shared concepts: **generative models**, **MLE**, **overfitting**, **probabilistic inference**
>* Difference:
>  * Regression: analytic solutions available in many cases
>  * Classification: no analytic solutions available $\rightarrow$ use **optimisation** for MLE & **approximation** for probabilistic inference

* **Classification Problems**:

>|  Application | Inputs | Outputs | 
| :---: | :---: | :---: |
| email spam filtering | email message | spam / not spam |
| medical diagnosis | patient medical data | disease diagnosis |
| object recognition | image RGB pixel values | object present in scene |  
| speech recognition | audio waveform | words spoken | 
| machine translation | text in language 1 | text in language 2 |

* **Subclasses of Classification Problem**:

>|Subclass|Explanation|Example|
|-|-|-|
|**Binary**     |two possible output values|spam filtering, medical diagnosis|
|**Multi-class**|many possible output values|object recognition with single image|
|**Multi-label**|multi-dimensional outputs (i.e. multiple labels per input)|object recognition with many objects in each image|
|**Structured** |Multi-label tasks w/ strong dependencies between output dimensions|machine translation|
