# 1. 2D Fourier Transforms and Linear Systems

* integral: from $-\infty$ to $+\infty$ unless specified otherwise

## 1.1. 2D Fourier Transform

* **1D FT**

>\begin{align}
G(\omega) = &\int^{+\infty}_{-\infty} g(t) e^{-j \omega t} dt \\
g(t) = \frac{1}{2\pi} &\int^{+\infty}_{-\infty} G(\omega) e^{j \omega t} d\omega
\end{align}

* **2D FT**

>\begin{align}
G(\omega_1, \omega_2) = &\int^{+\infty}_{-\infty} \int^{+\infty}_{-\infty} g(u_1,u_2) e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \\
g(u_1, u_2) = \frac{1}{(2\pi)^2} &\int^{+\infty}_{-\infty} \int^{+\infty}_{-\infty} G(\omega_1, \omega_2) e^{j(\omega_1 u_1 + \omega_2 u_2)} d\omega_1 d\omega_2
\end{align} 

* **2D Inverse FT: Proof** (integral: from $-\infty$ to $+\infty$)

>\begin{align}
FT(g(u_1,u_2)) &= \frac{1}{(2\pi)^2} \iint \left[ \iint G(\omega'_1, \omega'_2) e^{j(\omega'_1 u_1 + \omega'_2 u_2)} d\omega'_1 d\omega'_2 \right] e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \\
&= \frac{1}{(2\pi)^2} \iint G(\omega'_1, \omega'_2) \left[ \iint e^{j((\omega'_1 - \omega_1) u_1 + (\omega'_2 - \omega_2) u_2)} du_1 du_2 \right] d\omega'_1 d\omega'_2 \\
&= \frac{1}{(2\pi)^2} \iint (2\pi)^2 G(\omega'_1, \omega'_2) \delta(\omega'_1-\omega_1) \delta(\omega'_2-\omega_2) d\omega'_1 d\omega'_2 \\
&= G(\omega_1,\omega_2)
\end{align}

>* **NOTE:** $\int^{+\infty}_{-\infty} \exp (\pm j\omega t) d\omega = 2\pi\delta (t)$

* **Separability of 2D FT**

>$$G(\omega_1, \omega_2) = \int^{+\infty}_{-\infty} e^{-j\omega_1 u_1} \left[ \int^{+\infty}_{-\infty} g(u_1,u_2) e^{-j\omega_2 u_2} du_2 \right] du_1$$

>* **Step 1:** FT along $u_2$ direction at some fixed $u_1$
>* **Step 2:** FT along $u_1$ direction

## 1.2. Properties of 2D FT

* **Spatial Shift**

>\begin{align}
G'(\omega_1, \omega_2) &= \iint g(u_1 - \mu_1, u_2 - \mu_2) e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \\
&= \iint g(u'_1, u'_2) e^{-j[\omega_1 (u'_1 + \mu_1) + \omega_2 (u'_2 + \mu_2)]} du'_1 du'_2 \\
&= e^{-j(\mu_1 \omega_1 + \mu_2 \omega_2)} G(\omega_1, \omega_2)
\end{align}

>$$\boxed{\; g(u_1 - \mu_1, u_2 - \mu_2) \Leftrightarrow e^{-j(\mu_1 \omega_1 + \mu_2 \omega_2)} G(\omega_1, \omega_2) \;}$$

* **Frequency Shift** (or **Spatial Modulation**)

>\begin{align}
G(\omega_1 - \Omega_1, \omega_2 - \Omega_2) &= \iint g(u_1,u_2) e^{-j[(\omega_1 - \Omega_2)u_1 + (\omega_2 - \Omega_2)u_2]} du_1 du_2 \\
&= \iint e^{j(\Omega_1 u_1 + \Omega_2 u_2)} g(u_1,u_2) e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2
\end{align}

>$$\boxed{\; g(u_1,u_2) e^{j(\Omega_1 u_1 + \Omega_2 u_2)} \Leftrightarrow G(\omega_1 - \Omega_1, \omega_2 - \Omega_2) \;}$$

## 1.3. Useful FT

* **Impulse Function**

>\begin{align}
\text{Unit Area: } &\lim_{\epsilon \rightarrow 0} \int^\epsilon_{-\epsilon} \int^\epsilon_{-\epsilon} \delta(u_1, u_2) du_1 du_2 = 1\\
\text{Shifting Property: } &\iint \delta(u_1 - \mu_1, u_2 - \mu_2) f(u_1,u_2) du_1 du_2 = f(\mu_1,\mu_2) \\
\text{FT: } &\iint \delta(u_1, u_2) e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 = 1
\end{align}

>$$\boxed{\; \delta(u_1,u_2) \Leftrightarrow 1 \;}$$

* **Constant Level**

>$$\text{useful identity: } \lim_{A \rightarrow \infty} \int^A_{-A} \exp (\pm j\omega t) d\omega = 2\pi \delta (t)$$

>\begin{align}
G(\omega_1, \omega_2) &= \iint k\; e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \\
&= k(2\pi\delta(\omega_1))(2\pi\delta(\omega_2)) = k(2\pi)^2 \delta(\omega_1, \omega_2)
\end{align}

>$$\boxed{\; k \Leftrightarrow k(2\pi)^2\delta(\omega_1, \omega_2) \;}$$

* **Complex Exponential**

>\begin{align}
G(\omega_1, \omega_2) &= \iint e^{j(\Omega_1 u_1 + \Omega_2 u_2)} e^{-j (\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \\
&= \iint e^{-j[(\omega_1 - \Omega_1)u_1 + (\omega_2 - \Omega_2)u_2]} du_1 du_2 \\
&= (2\pi)^2 \delta(\omega_1 - \Omega_1, \omega_2 - \Omega_2)
\end{align}

>$$\boxed{\; e^{j(\Omega_1 u_1 + \Omega_2 u_2)} \Leftrightarrow (2\pi)^2 \delta (\omega_1 - \Omega_1, \omega_2 - \Omega_2) \;}$$

>* Or, **frequency shift theorem** $\rightarrow$ set $g(u_1,u_2)=1$ and $G(\omega_1, \omega_2) = (2\pi)^2 \delta(\omega_1,\omega_2)$

## 1.4. Linear Systems

* **Definition**

>$$\mathcal{F} [ax_1(u_1,u_2)+bx_2(u_1,u_2)] = a\mathcal{F}[x_1(u_1,u_2)] + b\mathcal{F}[x_2(u_1,u_2)]$$

* **Impulse Response**

>$$h(u_1,u_2 ; \mu_1,\mu_2) = \mathcal{F} [\delta(u_1 - \mu_1, u_2 - \mu_2)]$$

* **Spatial Invariance**

>* Shape of the impulse response is independent of the position of the impulse function

>$$\mathcal{F}[\delta(u_1-\mu_1,u_2-\mu_2)] = h(u_1-\mu_1,u_2-\mu_2)$$
>$$$$
>$$\text{where, } h(u_1,u_2) = \mathcal{F}[\delta(u_1,u_2)]$$

>* **Region of support $R_h$**: IR is zero outside $R_h$

>  * **IIR**(Infinite IR) & **FIR**(Finite IR) system: $R_h$ is infinite/finite
>* If the input 2d signal is greylevel image $\rightarrow$ IR is referred to as the **point spread function**

* **Convolution** for a linear spatially invariant (LSI) system, $\delta(\cdot)$ produces $h(\cdot)$

>\begin{align}
x(u_1,u_2) &= \iint x(u'_1,u'_2) \delta(u_1-u'_1,u_2-u'_2) du'_1 du'_2 \\
y(u_1,u_2) &= \iint x(u'_1,u'_2) h(u_1-u'_1,u_2-u'_2) du'_1 du'_2 \\
&= \iint h(u'_1,u'_2) x(u_1-u'_1,u_2-u'_2) du'_1 du'_2
\end{align}

* **Convolution Theorem**, $H(\omega_1,\omega_2)$: frequency response

>\begin{align}
Y(\omega_1,\omega_2) &= \iint \left[ \iint h(u'_1,u'_2) x(u_1-u'_1,u_2-u'_2) du'_1 du'_2 \right] e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \\
&= \iint h(u'_1,u'_2) \left[ \iint x(u_1-u'_1,u_2-u'_2) e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \right] du'_1 du'_2 \\
&= \iint h(u'_1,u'_2) \left[ e^{-j(\omega_1 u'_1 + \omega_2 u'_2)} X(\omega_1, \omega_2) \right] du'_1 du'_2
\end{align}

>$$\boxed{\; Y(\omega_1,\omega_2) = X(\omega_1,\omega_2) H(\omega_1,\omega_2)\;}$$

>* More generally, 

>$$\boxed{\; y(u_1,u_2) = f(u_1,u_2)*g(u_1,u_2) \rightarrow Y(\omega_1,\omega_2)=F(\omega_1,\omega_2)G(\omega_1,\omega_2) \;}$$

# 2. Image Digitisation and 2D DFT

## 2.1. Image Digitisation

* **Steps in Image Digitisation**

>* **Sampling:** convert image into a finite set of sample values
>  * Rectangular Grid
>  * Diamond Grid - When the image does not contain high freq. simultaneously in both dimensions
>  * Hexagonal Grid - For functions that are circularly symmetric or bandlimited over a circular region
>* **Quantisation:** represent each sample value as a finite precision number

* **Model Digitized Images as Bandlimited Signals**

>\begin{align}
F(\omega_1, \omega_2) &= 0 \;\;\;\text{if}\;\;\; 
|\omega_1|>\omega_{1c} \;\;\;\text{and}\;\;\; |\omega_2|>\omega_{2c} \\
\text{for circularly symmetric FT: } F(\omega_1, \omega_2) &= 0 \;\;\;\text{if}\;\;\; \sqrt{\omega_1^2 + \omega_2^2} > \omega_{c}
\end{align}

>* $\omega_c$: bandwidth

## 2.2. Rectangular Sampling

* **Sampling Function**

>\begin{align}
s(u_1,u_2) &= \sum^\infty_{n_1 = -\infty} \sum^\infty_{n_2 = -\infty} \delta(u_1-n_1 \Delta_1, u_2 - n_2\Delta_2) \\
\end{align}

* **Fourier Coefficient**

>* **Recall** Fourier Series for 1D Periodic Fn.

>\begin{align}
f(x) &= \sum^\infty_{n=-\infty} c_n e^{jn\omega_0 x} \;\;\;\text{with}\;\;\; \omega_0=\frac{2\pi}{T} \\
c_n &= \frac{1}{T} \int^{\alpha+T}_\alpha f(x) e^{-jn\omega_0 x} dx
\end{align}

>* **Fourier Series for $s(u_1,u_2)$**

>$$s(u_1,u_2) = \sum^\infty_{p_1 = -\infty} \sum^\infty_{p_2 = -\infty} c(p_1,p_2) e^{j(p_1 \Omega_1 u_1 + p_2 \Omega_2 u_2)} \;\;\;\text{with}\;\;\; \Omega_1 = \frac{2\pi}{\Delta_1} 
\;\;\;,\;\;\; \Omega_2 = \frac{2\pi}{\Delta_2}$$

>* **Fourier Coefficient** (integration interval: any interval covering a whole period)

>\begin{align}
c(p_1,p_2) &= \frac{1}{\Delta_1 \Delta_2} \int^{\frac{\Delta_2}{2}}_{-\frac{\Delta_2}{2}} \int^{\frac{\Delta_1}{2}}_{-\frac{\Delta_1}{2}} s(u_1,u_2) e^{-j(p_1 \Omega_1 u_1 + p_2 \Omega_2 u_2)} du_1 du_2 \\
&= \frac{1}{\Delta_1 \Delta_2} \int^{\frac{\Delta_2}{2}}_{-\frac{\Delta_2}{2}} \int^{\frac{\Delta_1}{2}}_{-\frac{\Delta_1}{2}} \left[ \sum^\infty_{n_1 = -\infty} \sum^\infty_{n_2 = -\infty} \delta(u_1-n_1 \Delta_1, u_2 - n_2\Delta_2) \right] e^{-j(p_1 \Omega_1 u_1 + p_2 \Omega_2 u_2)} du_1 du_2 \\
&= \frac{1}{\Delta_1 \Delta_2} \;\;\;\text{for all}\;\;\; p_1,p_2
\end{align}

* **Sampled Image**

>\begin{align}
g_s(u_1,u_2) &= g(u_1,u_2) s(u_1,u_2) \\
&= g(u_1,u_2) \frac{1}{\Delta_1 \Delta_2} \sum^\infty_{p_1 = -\infty} \sum^\infty_{p_2 = -\infty} e^{j(p_1 \Omega_1 u_1 + p_2 \Omega_2 u_2)}
\end{align}

* **FT (Spectrum) of the Sampled Image**

>* **Recall** Frequency Shift (or Spatial Modulation) Theorem

>$$g(u_1,u_2) e^{j(p_1 \Omega_1 u_1 + p_2 \Omega_2 u_2)} \Leftrightarrow G(\omega_1 - \Omega_1 p_1, \omega_2 - \Omega_2 p_2)$$

>* **FT of Sampled 2D Signal:** periodic repetition of the spectrum of the unsampled 2D signal

>$$G_s(\omega_1,\omega_2) = \frac{1}{\Delta_1 \Delta_2} \sum^\infty_{p_1 = -\infty} \sum^\infty_{p_2 = -\infty} G(\omega_1 - p_1 \Omega_1, \omega_2 - p_2 \Omega_2)$$

## 2.3. Recovering Original Image

* **Nyquist Frequency**

><img src = 'images/image1_01.png' width=300>

>* **2D Sampling Theorem:** If the image is **spatially bandlimited to** $\Omega_{B1}$ and $\Omega_{B2}$, the original continuous image can be recovered from the sampled image by ideal low-pass filtering at $\Omega_{B1}$, $\Omega_{B2}$ **if** the samples are taken such that 

>$$ \Omega_1 = \frac{2\pi}{\Delta_1}> 2\Omega_{B1} \;\;\;,\;\;\; 
\Omega_2 = \frac{2\pi}{\Delta_2}> 2\Omega_{B2}$$

>* $2\Omega_{B1}$ and $2\Omega_{B2}$: **2D Nyquist Frequencies**
>* **Sampling Below Nyquist Rates $\rightarrow$ Aliasing** 

* **Reconstruction of the Original Image**

>$$g(u_1,u_2) = \sum_n \sum_m g(n\Delta_1, m\Delta_2) \text{sinc} \frac{(u_1 - n\Delta_1)\pi}{\Delta_1} \text{sinc} \frac{(u_2 - n\Delta_2)\pi}{\Delta_2}$$

>* Provided a signal is sampled at or above its 2D Nyquist frequencies, it is possible to perfectly reconstruct the signal from its samples
>* $\text{sinc}$ functions: interpolate between samples

* **Derivation**

>\begin{align}
H(\omega_1,\omega_2) &= \Delta_1 \Delta_2 \;\;\;\text{if}\;\;\; \omega_1,\omega_2 \in R \\
\\
G(\omega_1,\omega_2) &= H(\omega_1,\omega_2) G_s(\omega_1,\omega_2) 
\end{align}

>* $H$: low-pass filter
>* $R$: area in the frequency plane, completely containing the spectrum of the unsampled signal $\rightarrow$ commonly set as $[-\frac{\pi}{\Delta_i} < \omega_i < \frac{\pi}{\Delta_i}]$, since $\frac{\pi}{\Delta_i} > \Omega_{Bi}$
>* **Multiplication in spatial domain** implies **convolution in frequency domain**, and vice versa

>\begin{align}
g(u_1,u_2) &= \int^{\infty}_{-\infty} \int^{\infty}_{-\infty} h(u_1-u'_1,u_2-u'_2) g_s(u'_1,u'_2) du'_1 du'_2 \\
\\
g_s(u'_1,u'_2) &= \sum_n \sum_m g(n\Delta_1,m\Delta_2) \delta(u'_1 - n\Delta_1, u'_2 - m\Delta_2) \\
\\
g(u_1,u_2) &= \sum_n \sum_m \int^{\infty}_{-\infty} \int^{\infty}_{-\infty} h(u_1-u'_1,u_2-u'_2) g(n\Delta_1,m\Delta_2) \delta(u'_1 - n\Delta_1, u'_2 - m\Delta_2) du'_1 du'_2 \\
&= \sum_n \sum_m h(u_1-n\Delta_1,u_2-m\Delta_2) g(n\Delta_1,m\Delta_2) \\
\\
h(u_1,u_2) &= \text{sinc} \frac{u_1\pi}{\Delta_1} \text{sinc} \frac{u_2\pi}{\Delta_2} \\
\\
g(u_1,u_2) &= \sum_n \sum_m g(n\Delta_1, m\Delta_2) \text{sinc} \frac{(u_1 - n\Delta_1)\pi}{\Delta_1} \text{sinc} \frac{(u_2 - n\Delta_2)\pi}{\Delta_2}
\end{align}

## 2.4. Diamond Sampling

* **Sampling Functions**

><img src = 'images/image1_02.png' width=300>

>\begin{align}
s(u_1,u_2) &= s_1(u_1,u_2) + s_2(u_1,u_2) \\
s_1(u_1,u_2) &= \sum^\infty_{n_1 = -\infty} \sum^\infty_{n_2 = -\infty} \delta[u_1-2n_1\Delta_1,u_2-2n_2\Delta_2] \\
s_2(u_1,u_2) &= \sum^\infty_{n_1 = -\infty} \sum^\infty_{n_2 = -\infty} \delta[u_1-(2n_1+1)\Delta_1,u_2-(2n_2+1)\Delta_2] \\
\end{align}

* **Fourier Coefficients**

>\begin{align}
s_1(u_1,u_2) &= \sum^\infty_{p_1 = -\infty} \sum^\infty_{p_2 = -\infty} c_1(p_1,p_2) e^{j \left( p_1\frac{\Omega_1}{2}u_1 + p_2\frac{\Omega_2}{2}u_2 \right) } \\
c_1(p_1,p_2) &= \frac{1}{4\Delta_1\Delta_2} \int^{\Delta_2}_{-\Delta_2} \int^{\Delta_1}_{-\Delta_1} s_1(u_1,u_2) e^{-j \left( p_1\frac{\Omega_1}{2}u_1 + p_2\frac{\Omega_2}{2}u_2 \right) } du_1 du_2 \\
&= \frac{1}{4\Delta_1\Delta_2} \sum^\infty_{n_1 = -\infty} \sum^\infty_{n_2 = -\infty} \int^{\Delta_2}_{-\Delta_2} \int^{\Delta_1}_{-\Delta_1} \delta[u_1-2n_1\Delta_1,u_2-2n_2\Delta_2] e^{-j \left( p_1\frac{\Omega_1}{2}u_1 + p_2\frac{\Omega_2}{2}u_2 \right) } du_1 du_2 \\
&= \frac{1}{4\Delta_1\Delta_2}
\end{align}

>* Similarly,

>\begin{align}
s_2(u_1,u_2) &= \sum^\infty_{p_1 = -\infty} \sum^\infty_{p_2 = -\infty} c_2(p_1,p_2) e^{j \left( p_1\frac{\Omega_1}{2}u_1 + p_2\frac{\Omega_2}{2}u_2 \right) } \\
c_2(p_1,p_2) &= \frac{1}{4\Delta_1\Delta_2} \int^{\Delta_2}_{-\Delta_2} \int^{\Delta_1}_{-\Delta_1} s_2(u_1,u_2) e^{-j \left( p_1\frac{\Omega_1}{2}u_1 + p_2\frac{\Omega_2}{2}u_2 \right) } du_1 du_2 \\
&= \frac{1}{4\Delta_1\Delta_2} \sum^\infty_{n_1 = -\infty} \sum^\infty_{n_2 = -\infty} \int^{2\Delta_2}_{0} \int^{2\Delta_1}_{0} \delta[u_1-(2n_1+1)\Delta_1,u_2-(2n_2+1)\Delta_2] e^{-j \left( p_1\frac{\Omega_1}{2}u_1 + p_2\frac{\Omega_2}{2}u_2 \right) } du_1 du_2 \\
&= \frac{1}{4\Delta_1\Delta_2} e^{-j \left( \frac{p_1 \Omega_1 \Delta_1}{2} + \frac{p_2 \Omega_2 \Delta_2}{2} \right) } = \frac{1}{4\Delta_1\Delta_2} e^{-j (p_1+p_2) \pi }
\end{align}

* **Sampled Image**

>\begin{align}
g_s(u_1,u_2) &= g(u_1,u_2) [s_1(u_1,u_2) + s_2(u_1,u_2)] \\
&= \sum^{\infty}_{p_1=-\infty} \sum^{\infty}_{p_2=-\infty} \big[ c_1(p_1,p_2)+c_2(p_1,p_2) \big] g(u_1,u_2) e^{j \left( p_1\frac{\Omega_1}{2}u_1 + p_2\frac{\Omega_2}{2}u_2 \right) }
\end{align}

* **FT (Spectrum) of the Sampled Image**

>\begin{align}
G_s(\omega_1,\omega_2) &= \sum^{\infty}_{p_1=-\infty} \sum^{\infty}_{p_2=-\infty} \big[ c_1(p_1,p_2)+c_2(p_1,p_2) \big] G \left( \omega_1 - p_1 \frac{\Omega_1}{2},\omega_2 - p_2 \frac{\Omega_2}{2} \right) \\
&= \frac{1}{4\Delta_1\Delta_2} \sum^{\infty}_{p_1=-\infty} \sum^{\infty}_{p_2=-\infty} 
\big[ 1 + e^{-j(p_1+p_2)\pi} \big] 
G \left( \omega_1 - p_1 \frac{\Omega_1}{2},\omega_2 - p_2 \frac{\Omega_2}{2} \right) \\
\\
1+e^{-j(p_1+p_2)\pi} &= \bigg\{ \begin{matrix} 0 && p_1+p_2=\text{odd} \\ 2 && p_1+p_2=\text{even} \end{matrix} \\
\\
G_s(\omega_1,\omega_2) &= \frac{1}{2\Delta_1\Delta_2} \sum^{\infty}_{p_1=-\infty} \sum^{\infty}_{p_2=-\infty} G \left( \omega_1 - p_1 \frac{\Omega_1}{2},\omega_2 - p_2 \frac{\Omega_2}{2} \right) \;\;\;\text{for}\;\;\; p_1+p_2=\text{even}\\
\end{align}


## 2.5. 2D Discrete Fourier Transform

* **Goal:** finite 2D array of numbers $\rightarrow$ approximate FT, given by a finite array of the same size (fully invertible)

* **Image Quantisation**

>* Eye is significantly less sensitive to quantisation error in images
>* **Audio:** 8 bits for general purpose / 16 bits for high quality
>* **Image:** 8 bits for general purpose / 10 or 12 bits for high quality

* **1D DFT**

>\begin{align}
F_k &= F(k\omega) = \sum^{N-1}_{n=0} f_n e^{-j\pi\frac{nk}{N}} \;\;\; &0 \leq k \leq N-1\\
f_n &= f(nT_s) = \frac{1}{N} \sum^{N-1}_{k=0} F_k e^{j\pi\frac{nk}{N}} \;\;\; &0 \leq n \leq N-1
\end{align}

>* $\omega_0 = 2\pi/T$ and $T=NT_s$

* **2D DFT** (separable to two 1D DFTs)

>\begin{align}
F_{p_1,p_2} &= \sum^{N_2-1}_{n_2=0} \sum^{N_1-1}_{n_1=0} f_{n_1,n_2} e^{-2\pi j\left(p_1\frac{n_1}{N_1} + p_2\frac{n_2}{N_2} \right)} \\
f_{n_1,n_2} &= \frac{1}{N_1 N_2} \sum^{N_2-1}_{p_2=0} \sum^{N_1-1}_{p_1=0} F_{p_1,p_2} e^{2\pi j\left(n_1\frac{p_1}{N_1} + n_2\frac{p_2}{N_2} \right)}
\end{align}

>* $F_{p_1,p_2}$: FT at frequencies $(p_1\Omega_1, p_2\Omega_2)$
>* $f_{n_1,n_2}$: image values at $(n_1\Delta_1, n_2\Delta_2)$

* **Derivation**

>\begin{align}
f_s(u_1,u_2) &= \sum^{N_2-1}_0 \sum^{N_1-1}_0 f(u_1,u_2) \delta(u_1-n_1\Delta_1,u_2-n_2\Delta_2) \\
\bar{f_s}(u_1,u_2) &= \sum^\infty_{p_2=-\infty} \sum^\infty_{p_1=-\infty} c(p_1,p_2) e^{j(p_1 \Omega_1 u_1 + p_2 \Omega_2 u_2)} \\
c(p_1,p_2) &= \frac{1}{T_1 T_2} \int^{+T_2/2}_{-T_2/2} \int^{+T_1/2}_{-T_1/2} f_s (u_1,u_2) e^{-j(p_1 \Omega_1 u_1 + p_2 \Omega_2 u_2)} du_1 du_2 \\
&= \frac{1}{T_1 T_2} \sum^{N_2-1}_{n_2=0} \sum^{N_1-1}_{n_1=0} f(n_1\Delta_1, n_2\Delta_2) e^{-j(p_1 \Omega_1 n_1 \Delta_1 + p_2 \Omega_2 n_2 \Delta_2)}
\end{align}

>* $\bar{f_s}$: $f_s$ tiled in the whole 2D plane
>* $\Omega_i = 2\pi/T_i$ and $T_i = N_i \Delta_i$
>* $c(p_1,p_2) T_1 T_2$: can be associated with the FT at $(p_1 \Omega_1, p_2, \Omega_2)$

## 2.6. 2D DFT - Effects of Shifting

* **Shifting:** interchange 1st & 3rd and 2nd & 4th quadrants

><img src = 'images/image1_03.png' width=300>

>* Shifting puts the **DC level** at the centre of the image

* **Effects of Shifting**

>\begin{align}
f_{n_1 + \frac{N_1}{2}, n_2 + \frac{N_2}{2}} &= \frac{1}{N_1 N_2} \sum^{N_2 - 1}_{p_2 = 0} \sum^{N_1 - 1}_{p_1 = 0} F_{p_1,p_2} e^{2\pi j \left( \left[ n_1 + \frac{N_1}{2} \right] \frac{p_1}{N_1} + \left[ n_2 + \frac{N_2}{2} \right] \frac{p_2}{N_2}\right)} \\
&= \frac{1}{N_1 N_2} \sum^{N_2 - 1}_{p_2 = 0} \sum^{N_1 - 1}_{p_1 = 0}   
\underset{\text{phase factor}}{\left( F_{p_1,p_2} e^{\pi j p_1 + \pi j p_2} \right)} e^{2\pi j \left( n_1 \frac{p_1}{N_1} + n_2 \frac{p_2}{N_2}\right)} \\
\end{align}

>* Phase factor: $e^{\pi j p_i}$ changes from $+1$ to $-1$ moving from even to odd pixels


# 3. 2D Digital Filters

## 3.1. 2D Digital Filter

* **Filter Relations**

>\begin{align}
H(\omega_1,\omega_2) &= \sum^\infty_{n_2 = -\infty} \sum^\infty_{n_1 = -\infty} h(n_1,n_2) e^{-j(\omega_1 n_1 + \omega_2 n_2)} \\
h(n_1, n_2) &= \frac{1}{(2\pi)^2} \int^\pi_{-\pi} \int^\pi_{-\pi} H(\omega_1, \omega_2) e^{j(\omega_1 n_1 + \omega_2 n_2)} d\omega_1 d\omega_2
\end{align}

* **Derivation**

>\begin{align}
h_s(u_1,u_2) &= \sum^\infty_{n_2 -\infty} \sum^\infty_{n_1 = -\infty} h(u_1,u_2) \delta(u_1 - n_1\Delta_1, u_2 - n_2\Delta_2) \\
H_s(\omega_1,\omega_2) &= \iint^\infty_{-\infty} h_s(u_1,u_2) e^{-j(\omega_1 u_1 + \omega_2 u_2)} du_1 du_2 \\
&= \sum^\infty_{n_2 -\infty} \sum^\infty_{n_1 = -\infty} h(n_1\Delta_1, n_2\Delta_2) e^{-j(\omega_1 n_1 \Delta_1 + \omega_2 n_2 \Delta_2)}
\end{align}

>* $H_s(\omega_1,\omega_2)$: periodic with periods $\Omega_i = 2\pi/\Delta_i$, written in Fourier Series expansion
>* $h(n_1\Delta_1, n_2\Delta_2)$: Fourier coefficients

>$$h(n_1\Delta_1,n_2\Delta_2) = \frac{1}{\Omega_1\Omega_2} \int^{\Omega_2/2}_{-\Omega_2/2} \int^{\Omega_1/2}_{-\Omega_1/2} H_s(\omega_1,\omega_2) e^{j(\omega_1 n_1 \Delta_1 + \omega_2 n_2 \Delta_2)} d\omega_1 d\omega_2$$

>* Define **normalised frequencies** ($\omega'_i = \omega_i \Delta_i$) and obtain final result

## 3.2. Zero-phase Filter

* **Phase is important in images:** mess up the phase $\rightarrow$ edges lost
* **Zero-phase Filter:** purely real

>\begin{align}
H(\omega_1,\omega_2) &= H^*(\omega_1,\omega_2) \\
\underset{(n_1,n_2 \in R_h)}{\sum \sum} h(n_1,n_2) e^{-j(\omega_1 n_1 + \omega_2 n_2)} &= \underset{(n_1,n_2 \in R_h)}{\sum \sum} h^*(n_1,n_2) e^{+j(\omega_1 n_1 + \omega_2 n_2)} \\
h(n_1,n_2) &= h^*(-n_1,-n_2)
\end{align}

>* Usually only the filters with **real** impulse response coefficients are considered:

>$$h(n_1,n_2) = h(-n_1,-n_2)$$

## 3.3. Ideal 2D Filters

* **Ideal Rectangular Bandpass Filters** (assume working with normalised freq.)

><img src = 'images/image1_04.png' width=300>

>* $H = H_u - H_l$
>* $H_u$: $1$ if $|w_1|<\Omega_{U1}$ and $|w_2|<\Omega_{U2}$, $0$ otherwise
>* $H_l$: $1$ if $|w_1|<\Omega_{L1}$ and $|w_2|<\Omega_{L2}$, $0$ otherwise

>\begin{align}
h(n_1,n_2) &= \frac{1}{(2\pi)^2} \left[ 
\int^{\Omega_{U2}}_{-\Omega_{U2}} \int^{\Omega_{U1}}_{-\Omega_{U1}} 
e^{j(\omega_1 n_1 + \omega_2 n_2)} d\omega_1 d\omega_2 -
\int^{\Omega_{L2}}_{-\Omega_{L2}} \int^{\Omega_{L1}}_{-\Omega_{L1}} 
e^{j(\omega_1 n_1 + \omega_2 n_2)} d\omega_1 d\omega_2
\right] \\
&= \frac{1}{(2\pi)^2} \left[ \frac{2\sin(\Omega_{U2}n_2)}{n_2} \frac{2\sin(\Omega_{U1}n_1)}{n_1} - \frac{2\sin(\Omega_{L2}n_2)}{n_2} \frac{2\sin(\Omega_{L1}n_1)}{n_1} \right] \\
&= \frac{1}{\pi^2} \left[ 
\Omega_{U2} \Omega_{U1} \text{sinc}(\Omega_{U2}n_2) \text{sinc}(\Omega_{U1}n_1) -
\Omega_{L2} \Omega_{L1} \text{sinc}(\Omega_{L2}n_2) \text{sinc}(\Omega_{L1}n_1)
\right]
\end{align}

>* Use unnormalised freq. (assume $\Omega_{Ui} < \pi/\Delta_i$

>$$h(n_1\Delta_1,n_2\Delta_2) = \frac{\Delta_1\Delta_2}{\pi^2} \left[ 
\Omega_{U2} \Omega_{U1} \text{sinc}(\Omega_{U2}n_2\Delta_2) \text{sinc}(\Omega_{U1}n_1\Delta_1) -
\Omega_{L2} \Omega_{L1} \text{sinc}(\Omega_{L2}n_2\Delta_2) \text{sinc}(\Omega_{L1}n_1\Delta_1)
\right]$$


* **Alternative Ideal Rectangular Bandpass Filter**

><img src = 'images/image1_05.png' width=300>

>* $H(\omega_1,\omega_2) = H_1(\omega_1) H_2(\omega_2)$
>* $H_1(\omega_1)$: $1$ if $\Omega_{L1}<|w_1|<\Omega_{U1}$, $0$ otherwise
>* $H_2(\omega_2)$: $1$ if $\Omega_{L2}<|w_2|<\Omega_{U2}$, $0$ otherwise

>\begin{align}
h(n_1,n_2) &= \frac{1}{2\pi} \int^\pi_{-\pi} H_1(\omega_1) e^{j\omega_1 n_1} d\omega_1 \times
\frac{1}{2\pi} \int^\pi_{-\pi} H_2(\omega_2) e^{j\omega_2 n_2} d\omega_2 \\
&= \frac{1}{\pi} [\Omega_{U1} \text{sinc}(\Omega_{U1}n1) - \Omega_{L1} \text{sinc}(\Omega_{L1}n1)] \times
\frac{1}{\pi} [\Omega_{U2} \text{sinc}(\Omega_{U2}n2) - \Omega_{L2} \text{sinc}(\Omega_{L2}n2)]
\end{align}

>* Difference of products $\rightarrow$ Product of differences

* **Ideally Circularly Symmetric Bandpass Filter**

><img src = 'images/image1_06.png' width=300>

>* $H(\omega_1,\omega_2)$: $1$ if $\Omega_L < |\boldsymbol{\omega}| < \Omega_U$, $0$ otherwise
>* Use $j(\omega_1 n_1 + \omega_2 n_2) = jr[n_1 \cos(\theta) + n_2 \sin (\theta)] = jr \sqrt{(n^2_1+n^2_2)} \sin(\theta+\phi)$
>* Where $\tan(\phi) = n_1/n_2$

>\begin{align}
h(n_1,n_2) &= \frac{1}{(2\pi)^2} \int^\pi_{-\pi} \int^\pi_{-\pi} H(\omega_1,\omega_2) e^{j(\omega_1 n_1 + \omega_2 n_2)} d\omega_1 d\omega_2 \\
&= \frac{1}{(2\pi)^2} \int^{\Omega_U}_{\Omega_L} r \int^\pi_{-\pi} e^{jr\sqrt{(n^2_1+n^2_2)}\sin(\theta+\phi)} d\theta dr \\
&= \frac{1}{(2\pi)^2} \int^{\Omega_U}_{\Omega_L} r \int^\pi_{-\pi}
\cos \left[ r \sqrt{(n^2_1+n^2_2)} \sin(\theta+\phi) \right] d\theta dr \\
&= 2 \times \frac{1}{(2\pi)^2} \int^{\Omega_U}_{\Omega_L} r \int^\pi_0
\cos \left[ r \sqrt{(n^2_1+n^2_2)} \sin(\theta) \right] d\theta dr
\end{align}

>* **Bessel Function:**

>\begin{align}
J_0(x) &= \int^\pi_0 \cos(x \sin\theta) d\theta \\
xJ_1(x) &= \int x J_0(x) dx
\end{align}

>* **Simplified Expression** $\left( \beta \equiv \sqrt{(n^2_1+n^2_2)} \right)$

>\begin{align}
h(n_1,n_2) &= \frac{1}{2\pi} \int^{\Omega_U}_{\Omega_L} r J_0 (r \beta) dr \\
&= \frac{1}{2\pi\beta} \int^{\beta\Omega_U}_{\beta\Omega_L} x J_0(x) dx = \frac{1}{2\pi\beta} \left[ \Omega_U J_1(\beta \Omega_U) - \Omega_L J_1(\beta \Omega_L) \right]
\end{align}

>* IR: difference of two 1st order Bessel fn.
